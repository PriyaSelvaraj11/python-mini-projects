```python
# imports
from PyQt5.QtWidgets import *
from PyQt5.QtCore import QCoreApplication
import sys


class Window(QDialog):

    # constructor
    def __init__(self):
        super(Window, self).__init__()

        # setting window title
        self.setWindowTitle("Simple interest calculator")

        # setting geometry to the window
        self.setGeometry(100, 100, 300, 400)

        # creating a group box
        self.formGroupBox = QGroupBox("Please fill the details below.")

        # creating a line edit for principal amt
        self.principalAmtLineEdit = QLineEdit()
        # creating a line edit for interest rate
        self.interestRateLineEdit = QLineEdit()
        # creating a line edit for number of years
        self.yearsNumberLineEdit = QLineEdit()

        # calling the method that create the form
        self.createForm()

        # creating a dialog button for ok and cancel
        self.buttonBox = QDialogButtonBox(QDialogButtonBox.Cancel)
        calculateButton = QPushButton("Calculate", self)
        self.buttonBox.addButton(calculateButton, QDialogButtonBox.AcceptRole)

        # adding action when form is accepted
        self.buttonBox.accepted.connect(self.getRepaymentAmount)

        # adding action when form is rejected
        self.buttonBox.rejected.connect(self.reject)

        # adding result label text
        self.resultText = QLabel(self)
        self.resultText.setWordWrap(True)

        # creating a vertical layout
        mainLayout = QVBoxLayout()

        # adding form group box to the layout
        mainLayout.addWidget(self.formGroupBox)

        # adding result text to the layout
        mainLayout.addWidget(self.resultText)

        # adding button box to the layout
        mainLayout.addWidget(self.buttonBox)

        # setting lay out
        self.setLayout(mainLayout)

    # get repayment amount
    def getRepaymentAmount(self):
        principalAmt = float(self.principalAmtLineEdit.text())
        interest = float(self.interestRateLineEdit.text())
        yearsNumber = float(self.yearsNumberLineEdit.text())

        repaymentamt = principalAmt * (1 + ((interest/100)*yearsNumber))
        resultText = "The total amount accured from simple interest on principal amount of $ {0} at rate of {1}% per annum for {2} years is $ {3}".format(
            principalAmt, interest, yearsNumber, repaymentamt)

        self.resultText.setText(resultText)

    # creat form method

    def createForm(self):

        # creating a form layout
        layout = QFormLayout()

        # adding rows
        # for principal amount
        layout.addRow(QLabel("Principal amount($)"), self.principalAmtLineEdit)

        # for interest rate per year
        layout.addRow(QLabel("Rate(% per year)"), self.interestRateLineEdit)

        # for number of years
        layout.addRow(QLabel("No. of years"), self.yearsNumberLineEdit)

        # setting layout
        self.formGroupBox.setLayout(layout)


# main method
if __name__ == '__main__':
    app = QCoreApplication.instance()
    if app is None:
        app = QApplication(sys.argv)

    # create the instance of our Window
    window = Window()

    # showing the window
    window.show()

    # start the app
    sys.exit(app.exec_())
```
